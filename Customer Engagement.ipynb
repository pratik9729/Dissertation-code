```python

def read_files_in_folder(folder_path):
    dataframes = {}
    files = os.listdir(folder_path)
    for file in files:
        if file.endswith('.csv'):
            file_path = os.path.join(folder_path, file)
            df = pd.read_csv(file_path)
            dataframes[file] = df
        elif file.endswith('.xlsx'):
            file_path = os.path.join(folder_path, file)
            df = pd.read_excel(file_path)
            dataframes[file] = df
    return dataframes

folder_path = ("/content/drive/MyDrive/Dissertation/CSV Data/2.Person Data")
dataframes = read_files_in_folder(folder_path)
dataframes.keys()
```

```python
# Assuming you have a DataFrame named 'asb' with columns 'Date_Opened' and 'Date_Closed'
asb['Date_Opened'] = pd.to_datetime(asb['Date_Opened'])
asb['Date_Closed'] = pd.to_datetime(asb['Date_Closed'])

# 1. Case Duration
asb['Case_Duration'] = (asb['Date_Closed'] - asb['Date_Opened']).dt.days
asb["Case_Duration"].fillna(-1 , inplace=True)
average_duration = asb['Case_Duration'].mean()
print(f'Average Case Duration: {average_duration} days')

# 3. Trends Over Time - Cases Opened
asb['Year_Month_Opened'] = asb['Date_Opened'].dt.to_period('M')
opened_counts = asb['Year_Month_Opened'].value_counts().sort_index()

# Convert the index and values to numeric
opened_counts.index = opened_counts.index.to_timestamp()
opened_counts = opened_counts.astype(int)

plt.figure(figsize=(10, 6))
sns.lineplot(x=opened_counts.index, y=opened_counts.values)
plt.title('Cases Opened Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


# 4. Trends Over Time - Cases Closed
asb['Year_Month_Closed'] = asb['Date_Closed'].dt.to_period('M')
closed_counts = asb['Year_Month_Closed'].value_counts().sort_index()

# Convert the index and values to integers
closed_counts.index = closed_counts.index.to_timestamp()
closed_counts = closed_counts.astype(int)

plt.figure(figsize=(10, 6))
sns.lineplot(x=closed_counts.index, y=closed_counts.values)
plt.title('Cases Closed Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


# 5. SLA Compliance
sla_duration = pd.to_timedelta('5 days')  # Example SLA duration

# Create a Boolean Series indicating whether each case meets the SLA criteria
within_sla = (asb['Date_Closed'] - asb['Date_Opened']) <= sla_duration

```
